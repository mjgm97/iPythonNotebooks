## 3.1 Exploración de datos.

La exploración es una de las bases de la ciencia de datos y una de las primeras tareas a las que un científico de datos puede enfrentarse. Normalmente, exploramos los datos la primera vez que nos llegan, intentando comprender de qué se tratan, identificando posibles patrones, y viendo distribuciones estadísticas que pueden ser de ayuda para futuros análisis más en profundidad. Algunas de las preguntas que surgen y a las que damos solución mediante la exploración de datos son: ¿Cuántos registros hay? ¿Son suficientes para un análisis en profundidad? ¿Qué variables hay en los datos? ¿De qué tipo son? En caso de que haya valores nulos, ¿podemos sustituirlos? Entre muchas otras.

Vamos a empezar leyendo el fichero, como hemos hecho en ocasiones anteriores. En esta ocasión, el fichero no está separado por comas, si no que se usa como separador el carácter '\t'. ¿Cómo leerías entonces el fichero?

```python
# Importamos Pandas

import pandas as pd
```

```python
# Vemos que no se lee de manera adecuada si lo intentamos leer como siempre. Usa el argumento "sep"

dataFrame = pd.read_csv("recursosNotebook/employeesSesion3.tsv", sep = "\t")
dataFrame
```

Ya hemos aprendido como leer ficheros en ``CSV`` mediante la función correspondiente. Sin embargo, en ocasiones querremos también escribir dichos ficheros en disco, ya sea para guardarnos un subconjunto de los datos, una versión modificada, o un nuevo conjunto de datos que hayamos creado. 

Para escribir ficheros en ``CSV``, usaremos la función ``DataFrame.to_csv()``.

La función ``shape`` nos permite ver las dimensiones de nuestro dataset, incluyendo el número de filas y de columnas.

```python
dataFrame.shape
```

Como vemos, tenemos un conjunto de 1000 filas y 8 columnas.

También podemos ver el tipo de datos que tiene cada columna. Ya conocemos la función ``type``, y ahora vamos a ponerla en práctica con una fila de nuestro ``DataFrame``:

```python
for valor in dataFrame[0:1].values[0]:
    print("El valor " + str(valor) + " es de tipo: " + str(type(valor)))
```
